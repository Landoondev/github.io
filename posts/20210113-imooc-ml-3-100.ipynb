# 3-10 numpy中的比较和 Fancy Indexing

```python
import numpy as np
```

```python
x = np.arange(16)
x
```

```python
# 索引一个值
x[3]
```

```python
# 索引区间切片
x[3:9]
```

```python
# 切片步长
x[3:8:2]
```

```python
# 访问非等步长元素
[x[3], x[5], x[8]]
```

```python
# 这就是 Fancy Indexing
ind = [3, 5, 8]

x[ind]
```

```python
ind = np.array([[0, 2],
                [1, 3]])

x[ind]
```

```python
X = x.reshape(4, -1)
X
```

```python
row = np.array([0, 1, 2])
col = np.array([1, 2, 3])
# (0, 1) (1, 2) (2, 3)

X[row, col]
```

```python
X[0, col]
```

```python
X[:2, col]
```

```python
# 对索引为 0, 2, 3 的列感兴趣 
col = [True, False, True, True]
```

```python
X[1:3, col]
```

## 逻辑运算

```python
x
```

```python
np.sum((x > 3) & (x < 10))
```

```python
np.sum((x % 2 == 0) | (x > 10))
```

```python
np.sum(~(x == 0))
```

```python
x[x < 5]
```

```python
x[x % 2 == 0]
```

```python
# 最后一列可以被 3 整除
X[X[:, 3] % 3 == 0, :]
```
